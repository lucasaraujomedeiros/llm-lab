```python
import requests
url = "https://www.gutenberg.org/files/100/100-0.txt"

def load_shakespeare():
    response = requests.get(url)
    response.raise_for_status()
    return response.text

text_data = load_shakespeare()
```

```python
import tiktoken
import torch

def text_to_token_ids(text, tokenizer):
    encoded = tokenizer.encode(text, allowed_special={'[EOS]'})
    encoded_tensor = torch.tensor(encoded).unsqueeze(0)
    return encoded_tensor
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
import re
words = re.split(r'([,.:;?_!"()\']|--|\s)', text_data)
words = [item.strip() for item in words if item.strip()]
```

```python
words_len = len(words)
tokens_len = len(text_to_token_ids(text_data, tokenizer)[0])
```

```python
print(f"Quantidade de tokens: {tokens_len}")
print(f"Quantidade de palavras: {words_len}")
print(f"Tokens por palavra: {tokens_len/words_len:.2f}")
```
